```python
import os
import pandas as pd
```

Apply rules to the affiliate_yt folder. Check any urls is labels as "others" by the rule

```python

affiliate_folder = '/home/data/chensun/affi_project/purl/output/affiliate_yt'
for crawl_id in os.listdir(affiliate_folder):
    each_crawl =  os.path.join(affiliate_folder, crawl_id)

    # Check if this crawl complete, or still building
    label_path = os.path.join(each_crawl, 'label.csv')
    if not os.path.exists(label_path):
        print("This crawl is not complete. Ignore")
        continue
    
    print("Processing file: ", each_crawl)

    rule_based_label_path = os.path.join(each_crawl, 'rule_based_label.csv')

    df_rule_based_label = pd.read_csv(rule_based_label_path, on_bad_lines='skip')
    df_rule_based_label['label'] = 'affiliate'


    df_rule_label_mismatch = df_rule_based_label[(df_rule_based_label['label'] == 'affiliate') & (df_rule_based_label['final_rules_based_label'] == 'others')]
    if len(df_rule_label_mismatch) == 0:
        print('\t\tNo mismatch')
    else:
        print("\t\tMismatch found")
        # Save the mismatch
        des_path = f"/home/data/chensun/affi_project/purl/output/rule_classifier_results/affiliate_yt/{crawl_id}"

        os.makedirs(des_path, exist_ok=True)
        rule_label_mismatch_path = os.path.join(des_path, 'rule_label_mismatch.csv')
        df_rule_label_mismatch.to_csv(rule_label_mismatch_path, index=False)

        print(f"Saved to {rule_label_mismatch_path}")
    
```
